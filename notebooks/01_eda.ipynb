## ***3. Data Visualization***

```python
class_counts = df2["Brand"].value_counts()
class_counts
```

```python
labels = class_counts.index
counts = class_counts.values
```

```python
plt.style.available
```

```python
sns.set_style("dark")
```

```python
plt.figure(figsize=(14, 6))

sns.barplot(x=labels, y=counts)

plt.xticks(rotation=45, ha='right')  # Rotate labels
plt.title("Brand Categories")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.grid(True)
plt.tight_layout()
plt.savefig("../plots/Brand_Categories.png", bbox_inches='tight')
plt.show()
```

### *Insights*
1. Maruti Suzuki dominates the dataset with more than 50k listings, indicating strong representation in the Indian used-car market.
2. Hyundai, Honda, Toyota, Mahindra, and Tata form the second tier of most common brands.
3. Luxury brands such as BMW, Audi, Jaguar, and Range Rover have significantly fewer samples.
4. The dataset is highly imbalanced across brands, which may cause the model to learn pricing patterns better for popular brands than luxury brands.

```python
fig, axes = plt.subplots(3,3,figsize=(22,11))
sns.histplot(ax=axes[0,0], data=df2, x="Car Type")
sns.histplot(ax=axes[0,1], data=df2, x="Transmission")
sns.histplot(ax=axes[0,2], data=df2, x="Fuel Type")
sns.histplot(ax=axes[1,0], data=df2, x="Kilometers", kde=True)
sns.histplot(ax=axes[1,1], data=df2, x="Year", kde=True)
sns.histplot(ax=axes[1,2], data=df2, x="Owner")
sns.histplot(ax=axes[2,0], data=df2, x="Accidental")
sns.histplot(ax=axes[2,1], data=df2, x="Price", kde=True)
sns.histplot(ax=axes[2,2], data=df2, x="Brand")
plt.xticks(rotation=45, ha='right')  # Rotate labels
plt.suptitle("Feature Distribution")
plt.tight_layout()
plt.savefig("../plots/feature_distribution.png")
plt.show()
```

### *Insights*
#### 1. Car Type
- The Indian used-car market remains heavily driven by affordable hatchback vehicles.

- Transmission is likely a significant feature affecting price.

- Electric vehicles remain underrepresented in the used-car market.
#### 4. Kilometers
- The model will be exposed to vehicles across a wide range of usage conditions.
#### 5. Year

```python
num_cols = df2.select_dtypes(include="number").columns
correlation_mat = df2[num_cols].corr()
sns.heatmap(
    correlation_mat,
    annot=True,
    fmt=".2f",
    cmap="coolwarm"
)
plt.title("Correlation Heatmap")
plt.savefig("../plots/correlation_heatmap.png")
plt.show()
```

### *Insights* 
1. Positive Correlation(0.41) between Year and Price - Newer vehicles tend to be more expensive.
2. Kilometers vs Price (-0.21) shows a weak negative relationship - Cars with higher mileage generally sell for lower prices.
3. Since correlations are relatively weak - This suggests tree-based models such as XGBoost, LightGBM,Random Forest will outperform simple Linear Models.

```python
 #Price trend over the years
sns.lineplot(
    data=df2,
    x="Year",
    y="Price",
    errorbar=None
)
plt.title("Price over the Years")
sns.set_style("dark")
plt.grid(True)
plt.savefig("../plots/price_trend.png")
plt.show()
```

### *Insights*
1. Average vehicle price generally increases with manufacturing year.
2. Significant jump after 2015.
3. Recency has a strong impact on resale value.
4. Year should be treated as a key predictive feature.

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df2,x="Kilometers",y="Price",hue="Fuel Type",alpha=0.6)
plt.title("Kilometers vs Price — where outliers live")
plt.tight_layout()
plt.savefig("../plots/km_vs_price_scatter.png")
plt.show()
```

### *Insights*
1. Higher mileage generally corresponds to lower vehicle prices.
2. The relationship is not strictly linear.
3. Mileage alone cannot explain price Brand, year, fuel type, ownership, and accident history must also be considered.
4. Potential anomalies can be identified where:
- High mileage cars are listed at unusually high prices.
- Low mileage cars are listed at suspiciously low prices.
> These will become candidates for your fraud detection system.

```python
sns.boxplot(data=df2, x="Brand", y="Price")
plt.xticks(rotation=45, ha='right')  # Rotate labels
plt.title("Outliers in Brand")
plt.tight_layout()
plt.savefig("../plots/outilers.png")
plt.show()
```

### *Insights*
1. Luxury-oriented brands such as MG, Kia, Range Rover, and Audi have higher median prices.
2. Mainstream brands such as Maruti, Hyundai, and Tata exhibit lower median prices.
3. Every brand contains several high-price outliers.
4. The extreme points above boxplot whiskers are natural candidates for anomaly detection.

---
## ***4. EDA Conclusion***

The EDA reveals that vehicle price is influenced by a combination of manufacturing year, mileage, brand, ownership history, transmission type, fuel type, and accident history. The target variable exhibits significant skewness and contains numerous outliers, making tree-based ensemble models such as XGBoost and LightGBM suitable choices. Furthermore, the presence of pricing anomalies across brands and mileage ranges validates the feasibility of building CarScope AI, a system capable of estimating fair market value and identifying suspiciously overpriced or underpriced vehicle listings. This directly supports the project's fraud detection objective.
